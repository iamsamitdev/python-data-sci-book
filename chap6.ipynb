```python
import pandas as pd

data = {'title':['Data Science', 'Database', 
                 'Big Data', 'Python Data Analytics'],
        
        'price': [500, 350, 450, 400]}

df = pd.DataFrame(data)


df1 = df[df.title.str.startswith('Data')]
df2 = df[df.title.str.endswith('Data')]
df3 = df[df['title'].str.find('Data') != -1]

display(df1, df2, df3)
```

```python
import pandas as pd

data = {'title':['Data Science', 'Database', 
                 'Big Data', 'Python Data Analytics'],
        
        'price': [500, 350, 450, 400]}

df1 = pd.DataFrame(data)
df1.title = df1.title.str.upper()

df2 = pd.DataFrame(data)
df2['title'] = df2.title.str.replace('Data', 'Info')

df3 = pd.DataFrame(data)
df3['title'] = df3['title'].str.strip(' ')

display(df1, df2, df3)
```
