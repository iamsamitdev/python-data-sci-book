```python
import pandas as pd
import numpy as np

def negative_bg(val):
    bg = '#ffc' if val < 0 else 'inherit'  #inherit คือใช้รูปแบบเดิม
    return f'background-color: {bg}'

data = {'One': np.random.randint(-5, 10, 10), 
        'Two': np.random.randint(-10, 100, 10),
        'Three': np.random.randint(-100, 1000, 10)}

df = pd.DataFrame(data)

df.style.applymap(negative_bg)
```

```python
import pandas as pd
import numpy as np

def negative_bg(val):
    bg = 'lightyellow' if val < 0 else 'inherit'
    return f'background-color: {bg}'

def negative_color(val):
    c = 'red' if val < 0 else 'inherit'
    return f'color: {c}'    

data = {'One': np.random.randint(-5, 10, 10), 
        'Two': np.random.randint(-10, 10, 10),
        'Three': np.random.randint(-100, 100, 10)}

df = pd.DataFrame(data)

df.style.applymap(negative_bg).applymap(negative_color)
```

```python
import pandas as pd
import numpy as np

data = {'One': np.random.randint(1, 10, 5), 
        'Two': np.random.randint(10, 100, 5),
        'Three': np.random.randint(100, 1000, 5)}

df = pd.DataFrame(data)
df.style.set_table_styles([
        {'selector': 'thead', 
         'props': [('border-bottom', 'none'), 
                   ('background-color', 'navy'), 
                   ('color', 'aqua'), ('font-size', '1.7rem')]},
    
        {'selector': 'tbody tr:hover', 
         'props': [('background-color', 'yellow !important')]},
    
        {'selector': 'tbody tr:nth-child(odd)', 
         'props': [('background-color', 'lightblue')]},
    
        {'selector': 'tbody tr:nth-child(even)', 
         'props': [('background-color', 'lightgreen')]},
    
        {'selector': 'tbody tr', 
         'props': [('border-top', 'solid 1px white')]}, 
    
        {'selector': 'td', 
         'props': [('padding', '10px'), ('font-size', '1.5rem')]}  
     ])
```

```python
import pandas as pd
import numpy as np

def negative_bg(val):
    bg = 'yellow' if val < 0 else 'white'
    return f'background-color: {bg}'

def negative_color(val):
    c = 'red' if val < 0 else 'black'
    return f'color: {c}'    

data = {'One': np.random.randint(-5, 10, 10), 
        'Two': np.random.randint(-10, 10, 10),
        'Three': np.random.randint(-100, 100, 10)}

df = pd.DataFrame(data)

df.style \
    .applymap(negative_bg) \
    .applymap(negative_color) \
    .to_excel('c:\py\dsc\styling.xlsx') 
```
