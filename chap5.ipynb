```python
import pandas as pd

data = { 'price': [200, 300, 250, 500], 
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1] }

df = pd.DataFrame(data, index=['p1','p2','p3','p4'])

display(df[df['price'] >= 300],
        df[(df.price >= 200) & (df.price <= 300)],
        df[df['stock'] == 0],
        df[['price', 'stock']][df.stock > 0])
```

```python
import pandas as pd

data = { 'price': [200, 300, 250, 500], 
         'sales': [0, 30, 25, 50],
         'stock': [8, 0, 5, 1],
         'discount': [10, 20, 0, 30] }

df = pd.DataFrame(data, 
                  index=['p1','p2','p3','p4'])

df += 10
df['price'] += 20
df['discount'] = 0
df['discount']['p2':'p4'] = 15

display(df)

new_df = df
new_df['price'] -= new_df['discount']
new_df.columns = ['dis_price', 'sales', 'stock', 'discount']

display(new_df[['dis_price', 'sales']])
```

```python
import pandas as pd

data = { 'price': [200, 300, 250], 
         'sales': [0, 30, 25],
         'stock': [8, 0, 5],
         'discount': [10, 20, 0] }

df = pd.DataFrame(data, 
                  index=['p1','p2','p3'])
    
for row in df.itertuples():
    pr = row.price
    sl = '-' if row.sales == 0 else row.sales
    st = '-' if row.sales == 0 else row.sales
    ds = '-' if row.discount == 0 else row.discount
    
    print(row.Index, pr, sl, st, ds, sep='\t')
```
