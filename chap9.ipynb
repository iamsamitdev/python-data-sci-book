```python
import pandas as pd

data = [5, 15, 10, 15, 20, 15, 10, 5]
idx = ['a', 'b', 'a', 'c', 'c', 'd', 'b', 'c']  

sr = pd.Series(data, index=idx)
display(sr.sort_index())  #แสดงข้อมูลทั้งหมดเรียงตาม Index ให้ดูง่ายขึ้น

sr.groupby(idx).sum()     #ผลรวมข้อมูลในแต่ละกลุ่ม
```

```python
import pandas as pd

data = pd.read_csv('data-sci\\product.csv')
df = pd.DataFrame(data)

#จัดกลุ่มสินค้าทั้งหมด
display(df.groupby('name')[['sales']].sum())

#จัดกลุ่มเฉพาะสินค้าที่มีในสต๊อก
display(df[df['stock'] > 0].groupby('name')[['sales']].sum())

#จัดกลุ่มเฉพาะสินค้าที่ไม่มีในสต๊อก
groups = df[df['stock'] == 0].groupby('name')
for (_, g) in groups:
    display(g)
```

```python
import pandas as pd

def func(g):
    return g['group'].count() > 3

#----------------------------------------------------------

data = {'group': ['B', 'C', 'B', 'A', 'A', 'C', 'B', 'C', 'A', 'C'], 
        'name': ['Jim', 'John', 'Jack', 'Jane', 'Joe', 'Jake', 
                 'Jeff', 'Jill', 'James', 'Jude'],
        'score': [70, 60, 80, 50, 70, 50, 90, 70, 80, 70]}

df = pd.DataFrame(data).sort_values('group')

g1 = df.groupby('group').filter(func)

g2 = df.groupby('group').filter(lambda x: x['score'].sum() > 230)

display(g1, g2)
```
