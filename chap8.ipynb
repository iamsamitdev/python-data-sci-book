```python
import pandas as pd
import numpy as np

data1 = {'A': np.random.randint(10, 100, 5),
        'B': np.random.uniform(1, 10, 5),
        'C': np.random.rand(5),
        'D': np.random.choice([7, 11, 101, 108.1009, 0.357], 5)}

frame1 = pd.DataFrame(data1)

data2 = {'Col1': np.random.randint(10, 100, 5),
        'Col2': np.random.uniform(1, 10, 5),
        'Col3': np.random.rand(5)}

frame2 = pd.DataFrame(data2)


hdf = pd.HDFStore('data-sci\\hdf.h5') 
hdf['frame1'] = frame1 
hdf['frame2'] = frame2

df1 = hdf['frame1']
df2 = hdf['frame2']

display(df1, df2)
```

```python
import pandas as pd
import numpy as np

data1 = {'A': np.random.randint(10, 100, 5),
         'B': np.random.uniform(1, 10, 5),
         'C': np.random.rand(5),
         'D': np.random.choice([7, 11, 101, 108.1009, 0.357], 5)}

frame1 = pd.DataFrame(data1)

data2 = {'Col1': np.random.randint(10, 100, 5),
         'Col2': np.random.uniform(1, 10, 5),
         'Col3': np.random.rand(5)}

frame2 = pd.DataFrame(data2)

frame1.to_hdf('data-sci\\hdf5.h5', key='frame1')
frame2.to_hdf('data-sci\\hdf5.h5', key='frame2')

df1 = pd.read_hdf('data-sci\\hdf5.h5', key='frame1')
df2 = pd.read_hdf('data-sci\\hdf5.h5', key='frame2')

display(df1, df2)
```

```python
import pandas as pd
import numpy as np

data = {'A': np.random.randint(10, 100, 10),
        'B': np.random.uniform(1, 10, 10),
        'C': np.random.rand(10),
        'D': np.random.choice([7, 11, 101, 108.1009, 0.357], 10)}

frame = pd.DataFrame(data)
frame.to_hdf('data-sci\\hdf_table.h5', 
              key='frame1', 
              format='table', 
              data_columns=['A', 'B', 'C'])


df = pd.read_hdf('data-sci\\hdf_table.h5', 
                  key='frame1', 
                  columns=['A', 'B', 'C'],
                  where='A >= 50 & B < 8')
df
```
